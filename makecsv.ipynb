```python
def placehold(x):
    x = x.replace("2011", 'twokeleven')
    x = x.replace("2001", 'twokone')
    x = x.replace("10 years", 'ten years')

    x = re.sub('[0-9]', 'x', x)

    x = x.replace("twokeleven", '2011')
    x = x.replace("twokone", '2001')
    x = x.replace("ten years", '10 years')


    x = x.replace("xx.x", "x.x")

    x = x.replace("\n", '')

    for area in areas:
        x = re.sub(area, '{placename}', x)

    for region in regions:
        x = re.sub(region, '{regionname}', x)

    return x

placehold("Kingston upon Hull")
```
